```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../Datasets/merged.csv")
```

```python
df.head()
```

```python
df.corr()
```

```python
df['AverageBrent'] = df['HighBrent'] + df['LowBrent'] / 2
df['AverageHeating'] = df['HighHeating'] + df['LowHeating'] / 2
df['AverageCrude'] = df['HighCrude'] + df['LowCrude'] / 2
```

```python
df.head()
```

```python
df.drop(['Date', 'VolumeBrent', 'VolumeHeating', 'VolumeCrude', 'OpenBrent', 'HighBrent', 'LowBrent', 'CloseBrent', 'OpenHeating', 'HighHeating', 'LowHeating', 'CloseHeating'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# dropping all columns related to Natural
df.drop(['OpenNatural', 'HighNatural', 'LowNatura', 'CloseNatural', 'VolumeNatural'], axis=1, inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(df.drop(['AverageBrent', 'AverageHeating'], axis=1), df[['AverageBrent', 'AverageHeating']], test_size=0.2, random_state=101)
```

```python
xTrain.head()
```

```python
yTrain.head()
```

```python
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
model1 = LinearRegression()
model2 = XGBRegressor()
model3 = RandomForestRegressor()
```

```python
model1.fit(xTrain, yTrain)
model2.fit(xTrain, yTrain)
model3.fit(xTrain, yTrain)
```

```python
yPred1 = model1.predict(xTest)
yPred2 = model2.predict(xTest)
yPred3 = model3.predict(xTest)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print("Linear Regression: ", mean_squared_error(yTest, yPred1))
print("XGBoost: ", mean_squared_error(yTest, yPred2))
print("Random Forest: ", mean_squared_error(yTest, yPred3))
```

```python
from sklearn.metrics import r2_score
```

```python
print("Linear Regression: ", r2_score(yTest, yPred1))
print("XGBoost: ", r2_score(yTest, yPred2))
print("Random Forest: ", r2_score(yTest, yPred3))
```
