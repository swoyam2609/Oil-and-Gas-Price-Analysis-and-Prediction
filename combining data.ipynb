```python
import pandas as pd

# Read the individual datasets
brent_df = pd.read_csv("brent.csv")
crude_df = pd.read_csv("crude.csv")
heating_df = pd.read_csv("heating.csv")
natural_df = pd.read_csv("natural.csv")
```

```python
merged_df = pd.merge(brent_df, crude_df, on='Date')
```

```python
merged_df
```

```python
merged_df = pd.merge(merged_df, heating_df, on='Date')
```

```python
merged_df
```

```python
merged_df = pd.merge(merged_df, natural_df, on='Date')
```

```python
merged_df.head()
```

```python
merged_df['Date'] = pd.to_datetime(merged_df['Date'])
```

```python
merged_df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(18, 5))
plt.plot(merged_df['Date'], merged_df['VolumeBrent'], label='Brent')
plt.plot(merged_df['Date'], merged_df['VolumeCrude'], label='Crude')
plt.plot(merged_df['Date'], merged_df['VolumeHeating'], label='Heating')
plt.plot(merged_df['Date'], merged_df['VolumeNatural'], label='Natural')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(18, 5))
plt.plot(merged_df['Date'], merged_df['OpenBrent'], label='Brent')
plt.plot(merged_df['Date'], merged_df['OpenCrude'], label='Crude')
plt.plot(merged_df['Date'], merged_df['OpenHeating'], label='Heating')
plt.plot(merged_df['Date'], merged_df['OpenNatural'], label='Natural')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(18, 5))
plt.plot(merged_df['Date'], merged_df['HighBrent'], label='Brent')
plt.plot(merged_df['Date'], merged_df['HighCrude'], label='Crude')
plt.plot(merged_df['Date'], merged_df['HighHeating'], label='Heating')
plt.plot(merged_df['Date'], merged_df['HighNatural'], label='Natural')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(18, 5))
plt.plot(merged_df['Date'], merged_df['CloseBrent'], label='Brent')
plt.plot(merged_df['Date'], merged_df['CloseCrude'], label='Crude')
plt.plot(merged_df['Date'], merged_df['CloseHeating'], label='Heating')
plt.plot(merged_df['Date'], merged_df['CloseNatural'], label='Natural')
plt.legend()
plt.show()
```
